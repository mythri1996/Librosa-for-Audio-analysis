Using Librosa

```python
import librosa
```

```python
audio='/content/mixkit-very-sick-man-coughing-2219.wav'
samples, sample_rate = librosa.load(audio, sr=None)
print(sample_rate),print(type(samples))
```

Using scipy

```python
from scipy.io import wavfile
sample_rate, samples = wavfile.read(audio)
```

```python
print(sample_rate),print(samples)
```

Visualiza the data

```python
import librosa.display
import matplotlib.pyplot as plt
plt.figure(figsize=(14,5))
librosa.display.waveplot(samples,sr=sample_rate)
```
